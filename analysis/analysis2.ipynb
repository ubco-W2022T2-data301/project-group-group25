```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv(r"C:\Users\somes\Downloads\project-group-group25\data\raw\insurance.csv")

pal = sns.cubehelix_palette(10, rot=-.25, light=.7)
plot = sns.FacetGrid(df, row="region", hue="region", aspect=10, height=1, palette=pal)

plot.map(sns.kdeplot, "charges",
      bw_adjust=.5, clip_on=False,
      fill=True, alpha=1, linewidth=1.5)
plot.map(sns.kdeplot, "charges", clip_on=False, color="w", lw=2, bw_adjust=.5)

plot.refline(y=0, linewidth=2, linestyle="-", color=None, clip_on=False)

def label(x, color, label):
    ax = plt.gca()
    ax.text(0, .2, label, fontweight="bold", color=color,
            ha="left", va="center", transform=ax.transAxes)

plot.map(label, "charges")

plot.figure.subplots_adjust(hspace=-.1)

plot.set_titles("")
plot.set(yticks=[], ylabel="")
plot.despine(bottom=True, left=True)

#From the following graph we can deduce that people who live in the northeast region were far likelier to be charged higher for their insurance price while the northwest region was the least likeliest.
```
